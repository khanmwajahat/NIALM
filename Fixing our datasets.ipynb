# Create a dataframe for every data file (190,195,200,205 etc). Either do all the train data first, or the test data.

### Things to do with each dataframe:

- ensure 127 number of columns <br>
- name columns <br>
- delete \r\r\n from i63 <br>
- insert a new column named class, with the correct class number for that combination <br>
- apply the clean_dataset function to remove NaNs <br>
- convert dataframe to type np.float just so everything is numeric <br>
- print df.head() or df.tail() to check what it looks like <br>

### Once you have all dataframes for that combination ready. You must:

-  Append them all <br>
-  Save to a new text file, name it *your combination*_train or *your combination*_test <br>
-  Read the new text file to ensure it is read correctly and everything looks good <br>

```python
import pandas as pd
import numpy as np
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```
